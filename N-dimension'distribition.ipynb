```python
import numpy as np
import math 
import matplotlib
import matplotlib.pyplot as plt
```

```python
def L1 (a,b):
    return np.sqrt(sum(np.power((a - b), 2)))
```

```python
arr = np.random.uniform(0,1,size=(1000,100))
```

```python
d = list()
```

```python
for i in range (0,999):
    c = i+1
    for j in range(c,1000):
        d.append( L1 (arr[i],arr[j]))
```

```python
plt.hist(d,bins=5000) 
```
