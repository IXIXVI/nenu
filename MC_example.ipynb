```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
aa=np.zeros((10000,21),int)
```

```python
def I(x):
    c=np.random.uniform()
    if x<=c:
        return -1
    else:
        return 1
```

```python
for i in range(10000):
    for j in range(0,20):
        aa[i][j+1]=aa[i][j]+I(0.5)
        if aa[i][j+1] < -10 or aa[i][j+1] > 10:
            aa[i][20]=100
            break
```

```python
num=0
for i in range(0,10000):
    if aa[i][20]==100:
        num+=1
```

```python
#存活率
1-(num/10000)
```

```python
b=np.zeros(21,int)
for j in range (10000):
    if aa[j][20] != 100:
        b[(aa[j][20])+10]+=1
b   
```

```python
c=np.arange(-10,11)
plt.xticks(c, color='blue',rotation=60)
plt.ylabel('num') 
plt.xlabel('situation') 
plt.bar(c,b,width=0.8)
```

```python
#n！：factorial
def fa(n):
    if n==2:
        return 2
    else:
        return n*fa(n-1)
#排列组合
def c(n,m):
    x=fa(n)/(fa(m)*fa(n-m))
    return x
```

```python
#真实存活概率
a=2*(np.power(0.5,11)+13*np.power(0.5,13)+c(15,2)*np.power(0.5,15)+c(17,3)*np.power(0.5,17)+c(19,4)*np.power(0.5,19))
1-a
```
